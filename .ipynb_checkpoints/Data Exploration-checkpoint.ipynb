```python
import pandas as pd
import time
import matplotlib.pyplot as plt
import seaborn as sns
import os
API_KEY =  os.environ['API_KEY']
%matplotlib inline
```

```python
from newsapi import NewsApiClient
```

```python
def get_data():
    try:
        newsapi = NewsApiClient(api_key = API_KEY)
        source = newsapi.get_sources(language = 'en')
        time.sleep(1)
        df = pd.DataFrame(columns=['id', 'name', 'description', 'url', 'category', 'countary'])
        for src in source['sources']:
            id_ = src['id']
            name = src['name']
            description = src['description']
            url = src['url']
            category = src['category']
            country = src['country']
            df = df.append({
                'id' : id_,
                'name' : name,
                'description' : description,
                'url' : url,
                'category' : category,
                'countary' : country
            },
                ignore_index=True)
        return df
    
    except Error as e:
        print(e)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import re
```

```python
def clean_txt(text):
    text = re.sub('[0-9]', '', text)
    text = "".join([char.lower() for char in text if char not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [word for word in tokens if word not in stopwords.words('english')]
    return text
```

```python
df_clean['description_clean'] = df_clean['description'].apply(lambda x : clean_txt(x))
```

```python
def db_connection(host, usr, passwd):
    try:
        conn = connect(host = host, user = usr, passwd = passwd, buffered = True)
        print('Connection is Successful!')
        return conn
    except Error as e:
        print(e)
    
```

```python
def select_db(db_name):
    try:
        cursor = conn.cursor()
    
        select_databases = f'USE {db_name};'
        
        print(f'{db_name} database selected succesfully!')

        databases = cursor.execute(select_databases)
    
        cursor.close()
        
    except Error as e:
        print(e)
    
   
```

```python
def show_databases():
    try:
        
        cursor = conn.cursor()
            
        show_databases = 'SHOW DATABASES;'

        cursor.execute(show_databases)
        
        databases = cursor.fetchall()
        
        for db in databases:
            print(f'{db[0]}')
    
        cursor.close()
    
    except Error as e:
        print(e)
 
    
```

```python
def create_database(db_name):
    try:
        
        cursor = conn.cursor()

        create_db = f'CREATE DATABASE IF NOT EXISTS {db_name};'
        
        print(f"'{db_name}' Database created successfully!")

        cursor.execute(create_db)
    
        cursor.close()
    except Error as e:
        print(e)
```

```python
create_database('new_data')
```

```python
def create_table(db_name, table_name, columns):
    try:
        
        cursor = conn.cursor()
        
        select_db(db_name)
    
        create_table = f'CREATE TABLE IF NOT EXISTS {table_name}({columns});'

        cursor.execute(create_table)
    
        print(f"'{table_name}' Table created successfully!")
    
        cursor.close()
    
    except Error as e:
        
        print(e)
    
```

```python
def show_tables(db_name):
    try:
        cursor = conn.cursor()
        
        select_db(db_name)
    
        show_tables = 'SHOW TABLES;'

        cursor.execute(show_tables)
        
        tables = cursor.fetchall()
        
        for tb in tables:
            print(f'{tb[0]}')
    
        cursor.close()
    
    except Error as e:
        print(e)
    
```
